```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
#load all 3 cleaned csv
csv_file = "Cleaned_CSVs/Symptoms.csv"
symptoms= pd.read_csv(csv_file)
symptoms.head()
```

```python
csv_file = "Cleaned_CSVs/VAERSVAX.csv"
vaersvax= pd.read_csv(csv_file)
vaersvax.head()
```

```python
csv_file = "Cleaned_CSVs/vaers_data.csv"
vaersdata= pd.read_csv(csv_file)
vaersdata.head()
```

```python
#performe the first merge
merged_outer_first = pd.merge(symptoms, vaersvax, on="VAERS_ID", how="outer")
merged_outer_first
```

```python
#perform the second merge
merged_outer_second = pd.merge(merged_outer_first, vaersdata, on="VAERS_ID", how="outer")
merged_outer_second
```

```python
#deleting unnessary columns
del merged_outer_second['Unnamed: 0_x']
```

```python
del merged_outer_second['Unnamed: 0_y']
```

```python
del merged_outer_second['Unnamed: 0']
```

```python
#overview of the dataframe
merged_outer_second
```

```python
#drop duplicates of any vares_id to prep the dataset (making sure the primary keys are unique)
merged_outer_second.drop_duplicates("VAERS_ID", inplace=True)
merged_outer_second
```

```python
#Cleanning the rows where vaccine column does not have any info (ie, dont know the brand of the vaccine)
merged_outer_second = merged_outer_second[merged_outer_second['Vaccine'].notna()]

merged_outer_second
```

## Loading data into SQL database

```python
#connecting to local host
rds_connection_string = "postgres:***@localhost:5432/covid_vaccine_db"
engine = create_engine(f'postgresql://{rds_connection_string}')
```

```python
#check for table names
engine.table_names()
```

```python
#load data into database
merged_outer_second.to_sql(name='covid_vaccine', con=engine, if_exists='append', index=False)
```

```python
#make sure it has been loaded correctly
pd.read_sql_query('select * from covid_vaccine', con=engine).head()
```
