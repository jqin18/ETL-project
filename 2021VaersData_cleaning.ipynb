```python
#import dependecies
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np


data_path = "2021VAERSDATA.csv"

vaers_data = pd.read_csv(data_path,low_memory=False)

vaers_data.head(10)
```

```python
#getting rid of columns that is not needed
vaers_data = vaers_data[["VAERS_ID", "RECVDATE", "STATE", "AGE_YRS", "SEX", "SYMPTOM_TEXT", "RECOVD", "ALLERGIES" ]]
vaers_data.head(10)
```

```python
#check the counts to each column
vaers_data.count()
```

```python
#found out 50% of the allergies column has none or NAN, not dropping it because it loses too much info
vaers_data['ALLERGIES'].value_counts(ascending = True)
```

```python
#will name most of the values NKA, which is a medical abbrevation for "not known allergies"
vaers_data["ALLERGIES"].replace({"NKDA":"NKA", "none":"NKA", "None":"NKA","na":"NKA"})
vaers_data["ALLERGIES"].fillna('NKA', inplace = True)
```

```python
#drop na
vaers_data = vaers_data.dropna(how='any')
```

```python
#new file
vaers_data.to_csv('vaers_data.csv', encoding='utf-8')
```
