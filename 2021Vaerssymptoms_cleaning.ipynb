```python
symptoms_df = pd.read_csv(symptoms_file)
symptoms_df
```

```python
symptoms_df.drop(columns=['SYMPTOMVERSION1', 'SYMPTOMVERSION2', 'SYMPTOMVERSION3','SYMPTOMVERSION4','SYMPTOMVERSION5'], inplace=True)
symptoms_df.drop_duplicates()
```

```python
symptoms_df.to_csv("ETL-project")
```

```python
engine = create_engine('sqlite://',
                       echo = False)
```

```python
symptoms_df.to_sql('Symptoms_Data',
               con = engine)
```
