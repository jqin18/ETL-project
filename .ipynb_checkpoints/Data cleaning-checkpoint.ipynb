```python
import pandas as pd
from sqlalchemy import create_engine
from config import password
engine = create_engine(f'postgresql://postgres:{password}@localhost:5432/employee_db')
```

```python
data_file = "./data/2021VAERSVAX.csv"
VAERSVAX_data_df = pd.read_csv(data_file)
VAERSVAX_data_df.head(20)
```

```python
# Create a filtered dataframe from specific columns
del VAERSVAX_data_df['VAX_LOT']
del VAERSVAX_data_df['VAX_NAME']
del VAERSVAX_data_df['VAX_TYPE']
VAERSVAX_data_df.head(100)
```

```python
VAERSVAX_data_df.dropna(how="any",inplace= True)
VAERSVAX_data_df.drop_duplicates()
VAERSVAX_data_df.head(100)
```

```python
VAERSVAX = VAERSVAX_data_df.rename(columns={'VAX_MANU':'Vaccine','VAX_DOSE_SERIES':'Dose','VAX_ROUTE':'Route','VAX_SITE':'Site'})
VAERSVAX
```

```python
VAERSVAX.to_csv('VAERSVAX.csv')
```
